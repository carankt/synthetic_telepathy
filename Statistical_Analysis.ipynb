```python
from dataloader import get_loader
root = '/Volumes/Datasets/inner_speech/derivatives/'
# root =  'dataset/derivatives/' # -sil
creater = get_loader(root)
xn, yn = creater.load_multiple_subjects([1, 2, 3, 4, 5, 6, 7, 8, 10])
```

```python
xn.shape
```

```python
labels = ["Up", "Down", "Right", "Left"] 
sorted_data = []
fs = 254
NFFT = 256
classes = yn[:,1]
for i in range(1,5):
    mask = classes == i
    data = xn[mask,:,:]
    sorted_data.append(data)
```

```python
class_data_stats = []
df_stats = []
for k in range(0, len(sorted_data)):
    data = sorted_data[k]
    #df_stats = pd.DataFrame()
    df_stats.append(pd.DataFrame(data.mean(axis = 0)).transpose().describe()) #

    
## Plot the Mean Values for each channel 
n = 4 # play with this and figsize to get better images
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(5*n, n))
for i in range(0,4):
    axs[i].plot(range(0, 128), df_stats[i].iloc[1, :])
    axs[i].set_title('Mean value across channels for Class: {}'.format(labels[i]))
    axs[i].set_xlabel('Channels')
    axs[i].set_ylabel('Values')


n = 4 # play with this and figsize to get better images
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(5*n, n))
for i in range(0,4):
    axs[i].plot(range(0, 128), df_stats[i].iloc[2, :])
    axs[i].set_title('Std value across channels for Class: {}'.format(labels[i]))
    axs[i].set_xlabel('Channels')
    axs[i].set_ylabel('Values')

    
n = 4 # play with this and figsize to get better images
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(5*n, n))
for i in range(0,4):
    axs[i].plot(range(0, 128), df_stats[i].iloc[3, :])
    axs[i].set_title('Min value across channels for Class: {}'.format(labels[i]))
    axs[i].set_xlabel('Channels')
    axs[i].set_ylabel('Values')


```
