```python
labels = ["Up", "Down", "Right", "Left"] 
sorted_data = []
fs = 254
NFFT = 256
classes = yn[:,1]
for i in range(1,5):
    mask = classes == i
    data = xn[mask,:,:]
    sorted_data.append(data)
```
